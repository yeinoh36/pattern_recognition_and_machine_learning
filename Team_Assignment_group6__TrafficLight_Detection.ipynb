```python
import yaml

# data.yaml 파일 경로
data_yaml_path = f"{dataset.location}/data.yaml"

# data.yaml 파일 로드
with open(data_yaml_path, 'r') as file:
     data = yaml.safe_load(file)

# 증강 설정 추가
data['degrees'] = 270.0

# data.yaml 파일 저장
with open(data_yaml_path, 'w') as file:
     yaml.safe_dump(data, file)
```
